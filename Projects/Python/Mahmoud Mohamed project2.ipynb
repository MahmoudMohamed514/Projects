```python

def plot_most_active_directors(df: pd.DataFrame, top_n: int = 20) -> None:
    """
    Plots a horizontal bar chart of the most active directors based on the number of movies.

    Args:
    df (pd.DataFrame): The DataFrame containing movie data with a 'director' column.
    top_n (int): The number of top directors to display. Default is 20.

    Returns:
    None: This function displays a plot and does not return any value.
    """
    plt.figure(figsize=(12, 8))
    df.director.value_counts().head(top_n).sort_values().plot(kind='barh', color='#FF4500', alpha=0.7)
    plt.title("Most Active Directors", fontsize=16, color='#FF4500', fontweight='bold')
    plt.xlabel("Number of Movies", fontsize=12, color='#FF4500')
    plt.ylabel("Director", fontsize=12, color='#FF4500')
    plt.xticks(fontsize=10, color='#FF4500')
    plt.yticks(fontsize=10, color='#FF4500')
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['left'].set_color('#FF4500')
    plt.gca().spines['bottom'].set_color('#FF4500')
    plt.show()
```

```python
# تحويل عمود genres إلى قائمة من الأنواع
df['genres'] = df['genres'].str.split('|')

# إنشاء جدول جديد يحتوي على متوسط الإيرادات لكل نوع في كل سنة
genre_year_revenue = df.explode('genres').groupby(['genres', 'release_year'])['revenue_adj'].mean().reset_index()

# رسم مخطط خطي يوضح التغيرات في متوسط الإيرادات لكل نوع على مر السنين
plt.figure(figsize=(18,14))
for genre in genre_year_revenue['genres'].unique():
    plt.plot(genre_year_revenue[genre_year_revenue['genres'] == genre]['release_year'],
             genre_year_revenue[genre_year_revenue['genres'] == genre]['revenue_adj'], label=genre)

plt.xlabel('Year')
plt.ylabel('Average Adjusted Revenue')
plt.title('Average Revenue by Genre Over Time')
plt.legend()
plt.show()
```

```python
# Calculate the average adjusted revenue for each director
director_revenue = df.groupby('director')['revenue_adj'].mean().reset_index()
director_revenue = director_revenue.sort_values('revenue_adj', ascending=False)

# Calculate the average adjusted revenue for each production company
production_company_revenue = df.explode('production_companies').groupby('production_companies')['revenue_adj'].mean().reset_index()
production_company_revenue = production_company_revenue.sort_values('revenue_adj', ascending=False)

# Print the top 10 directors and production companies by average adjusted revenue
print("Top 10 directors by average adjusted revenue:")
print(director_revenue.head(10))

print("\nTop 10 production companies by average adjusted revenue:")
print(production_company_revenue.head(10))

# Create a bar chart for theop 5 directors
plt.figure(figsize=(12, 6))
plt.bar(director_revenue['director'].head(5), director_revenue['revenue_adj'].head(5))
plt.xlabel('Director')
plt.ylabel('Average Adjusted Revenue')
plt.title('Top 5 Directors by Average Adjusted Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
# تحويل تاريخ  إلى نوع بيانات التاريخ
df['release_date'] = pd.to_datetime(df['release_date'])

df['release_month'] = df['release_date'].dt.month

# حساب متوسط الإيرادات بالشهر
monthly_revenue = df.groupby('release_month')['revenue_adj'].mean().reset_index()
plt.plot(monthly_revenue['release_month'], monthly_revenue['revenue_adj'])
plt.xlabel('Month')
plt.ylabel('Average Adjusted Revenue')
plt.title('Average Monthly Revenue')
plt.show()
```
